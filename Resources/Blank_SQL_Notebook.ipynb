```python
%%sql
select
   c.customerkey,
  p.productkey,
  p.productname,
  p.categoryname,
  orderdate,
  quantity*netprice*exchangerate as net_revenue,
  case
    when quantity*netprice*exchangerate >= 1000 then 'quality'
    else 'low quality'
  end as quality
from sales
left join customer c
  on sales.customerkey = c.customerkey
left join product p
  on sales.productkey = p.productkey
where orderdate ::DATE between '2020-01-01' and '2020-01-31'
order by net_revenue desc;
```

```python
%%sql
select
  orderdate,
  COUNT(distinct customerkey) as customers
from sales
where orderdate ::DATE between '2020-01-01' and '2020-01-31'
group by orderdate
order by customers desc;
```
